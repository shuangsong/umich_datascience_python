```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
#first we scrape belmont MA,data
#Loop through our URLs we loaded above

page = requests.get("http://www.city-data.com/city/Belmont-Massachusetts.html")
soup = BeautifulSoup(page.content, "html.parser")
#identify table we want to scrape

right_table=soup.find('table', {"class":'table tabBlue tblsort tblsticky sortable'})
rows = right_table.findAll("tr")

# header attributes of the table
header = [th.text.rstrip() for th in rows[0].find_all('th')]

# data
lst_data = []
for row in rows[1:]:
            data = [d.text.rstrip() for d in row.find_all('td')]
            lst_data.append(data)

# result
df_be = pd.DataFrame(lst_data, columns=header)
df_be.head()
```

```python
#i want to look at numbers of crimes(sum of those crimes) in those four cities and do data transformation:
#first, keep the data in the brackets and remove outside items:
#function of data transformation:
def my_func(df):
    for col in df.columns[1:]:
        df[col] = df[col].str.extract(r'(\w+).*\((\d+)', expand = False)
    return df

belmont = my_func(df_be)
belmont.head()
```
